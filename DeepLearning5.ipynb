```python
import pandas as pd
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv("pima-indians-diabetes.csv",header=None)
df
```

```python
x=df.iloc[:,0:8].values
y=df.iloc[:,8].values
```

```python
model=Sequential()
layers12=Dense(10,input_shape=(8,),activation="relu")
layers3=Dense(20,activation="relu")
layers4=Dense(1,activation="sigmoid")
model.add(layers12)
model.add(layers3)
model.add(layers4)
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
rajesh=model.fit(x,y,epochs=100,validation_split=0.33,batch_size=10,verbose=0)
rajesh.history["accuracy"]
```

```python
rajesh.history["loss"]
```

```python
plt.plot(rajesh.history["accuracy"])
plt.plot(rajesh.history["val_accuracy"])
plt.title("model loss")
plt.xlabel("accuracy")
plt.ylabel("loss")
plt.legend(["train","test"],loc="upper left")
```
